# Functions

```python
def word_tokenization_and_remove_punctuation(text):
    exclude = set(string.punctuation)
    tokenized = []
    tokenizer = TweetTokenizer()
    for doc in text:
        tokens = tokenizer.tokenize(doc.lower())
        tokenized.append(''.join([ch for ch in ' '.join(tokens) if ch not in exclude]).split())
    return tokenized
```

```python
def remove_stop_word(tokenized):
    sws = set(stopwords.words('english'))
    sws.add('rt') 
    sws.add("…") 
    sws_removed = []
    for j,sent in enumerate(tokenized):
        sws_removed.append([i for i in sent 
                            if i not in sws   
                                 and len(i) > 2  
                             and (not i.startswith("http"))])  
    return sws_removed
```

```python
def stem_sentence(text):
    
    word_tokens = word_tokenize(text)
    stemmed_words = [stemmer.stem(word) for word in word_tokens]
    return ' '.join(stemmed_words)
```

```python
def get_topic(ldamodel, document):
    """
    takes a trained ldamodel and a document (from the corpus)
    and returns the topic # it most closely associates with
    """
    topics = ldamodel.get_document_topics(document)
    topics = sorted(topics, key=lambda x: -x[1])
    return topics[0][0]



def topic_modeling(texts, df):
    try:
        
        dictionary = corpora.Dictionary(texts)
        dictionary.filter_extremes(no_below=4, no_above=0.3)
        dictionary.compactify()

        
        corpus = [dictionary.doc2bow(text) for text in texts]

        # Train the LDA model
        ldamodel = models.ldamodel.LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15, random_state=42)
        topics = [get_topic(ldamodel, doc) for doc in corpus]
        df['topic'] = topics

        # Get the topic distribution for each document
        topic_distributions = [ldamodel.get_document_topics(bow, minimum_probability=0) for bow in corpus]

        # Convert the list of distributions to a numpy array for easier handling later
        topic_distributions_np = np.array([[prob for _, prob in doc] for doc in topic_distributions])

        # Add topic distribution columns to df
        for i in range(topic_distributions_np.shape[1]):
            df[f'topic_{i}'] = topic_distributions_np[:, i]

        return ldamodel.print_topics(num_topics=10, num_words=3)
    except Exception as e:
        print("An error occurred:", e)
        return None
```

```python
def sentiment_analysis(df):
    sid = SentimentIntensityAnalyzer()
    scores = []


    for index, row in df.iterrows():

        sentiment_result = sid.polarity_scores(row['text'])
        scores.append(sentiment_result)


    df_scores = pd.DataFrame(scores)
    return df_scores
```

```python
def sentiment_analysis_on_each_topic(df):
    sid = SentimentIntensityAnalyzer()
    scores = []


    for index, row in df.iterrows():

        sentiment_result = sid.polarity_scores(row['text'])
        sentiment_result['topic'] = row['topic'] 
        scores.append(sentiment_result)


    df_scores = pd.DataFrame(scores)
    return df_scores
```

# Network Construction

```python

G = nx.Graph()

# Add user nodes from the reviews dataframe
for user in df['user_id'].unique():
    G.add_node(user, bipartite=0, type='user')

# Add magazine nodes from the metadata dataframe using 'parent_asin'
for magazine in meta_df['parent_asin'].unique():
    G.add_node(magazine, bipartite=1, type='magazine')

# Add edges between users and magazines based on reviews
for index, row in df.iterrows():

    if row['parent_asin'] in meta_df['parent_asin'].values:
        G.add_edge(row['user_id'], row['parent_asin'], rating=row['rating'])


```

```python
# Write the network into csv file to be used in Gephi
user_nodes = [n for n, d in G.nodes(data=True) if d['bipartite'] == 0]
magazine_nodes = [n for n, d in G.nodes(data=True) if d['bipartite'] == 1]

import csv


csv_file_path = 'network_graph.csv'


with open(csv_file_path, 'w', newline='') as csvfile:

    writer = csv.writer(csvfile)
    

    writer.writerow(['user_id', 'magazine_id', 'review_rating'])  
  
    for u, v, data in G.edges(data=True):
    
        writer.writerow([u, v, data.get('rating', '')]) 
```

```python
import random


random.seed(0)

# Sample a subset of user nodes
sampled_user_nodes = random.sample(user_nodes, 150)  


subgraph_nodes = sampled_user_nodes + [n for n in magazine_nodes if n in G]
subgraph = G.subgraph(subgraph_nodes)


sub_pos = nx.spring_layout(subgraph, k=0.15, iterations=20)


nx.draw_networkx_nodes(subgraph, sub_pos, nodelist=sampled_user_nodes, node_color='red', node_size=50, label='Sampled Users')
nx.draw_networkx_nodes(subgraph, sub_pos, nodelist=[n for n in subgraph_nodes if n in magazine_nodes], node_color='blue', node_size=5, label='Magazines')
nx.draw_networkx_edges(subgraph, sub_pos, width=1.0, alpha=0.5)


plt.axis('off')
plt.legend()
plt.show()
```

```python
# Detect communities in the graph
partition = community_louvain.best_partition(G)
# Store the community information in each node
for node, com_id in partition.items():
    G.nodes[node]['community'] = com_id
```

```python

# Calculate centrality for magazines to find the most influential magazines
centrality_scores = nx.degree_centrality(G)

# You might want to differentiate between user and magazine centralities
magazine_centrality = {node: centr for node, centr in centrality_scores.items() if G.nodes[node]['type'] == 'magazine'}
```

# Recommendation function setup

```python
import numpy as np

from sklearn.preprocessing import MinMaxScaler

# Create a scaler 
scaler = MinMaxScaler()


df['scaled_rating'] = scaler.fit_transform(df[['rating']])
df['scaled_helpful_votes'] = scaler.fit_transform(df[['helpful_vote']])
df['scaled_compound'] = scaler.fit_transform(df[['compound']])


# Scale centrality scores
centrality_values = list(magazine_centrality.values())
scaled_centrality = scaler.fit_transform(np.array(centrality_values).reshape(-1, 1))
magazine_centrality_scaled = dict(zip(magazine_centrality.keys(), scaled_centrality.flatten()))

# Scale topics
topic_columns = [f'topic_{i}' for i in range(10)]
topic_data = df[topic_columns]  # Extract topic data




scaled_topic_data = scaler.fit_transform(topic_data)

# Create new column names for the scaled topics
scaled_topic_columns = [f'scaled_{col}' for col in topic_columns]

# Assign scaled data back to the original DataFrame
for i, col_name in enumerate(scaled_topic_columns):
    df[col_name] = scaled_topic_data[:, i]

df
```

```python


def recommend_magazines(user_id, num_recommendations=5, weight_rating=0.3, weight_helpful_votes=0.2, weight_sentiment=0.2, weight_centrality=0.2, weight_topic=0.1, weight_community=0.1):
    user_reviews = df[df['user_id'] == user_id]
    reviewed_magazines = user_reviews['parent_asin'].tolist()
    user_community = G.nodes[user_id]['community']

    scores = {}
    for magazine in df['parent_asin'].unique():
        if magazine not in reviewed_magazines:
            magazine_entries = df[df['parent_asin'] == magazine]
            average_rating = magazine_entries['scaled_rating'].mean()
            total_helpful_votes = magazine_entries['scaled_helpful_votes'].sum()
            average_sentiment_score = magazine_entries['scaled_compound'].mean()
            centrality_score = magazine_centrality_scaled.get(magazine, 0)
            magazine_index = magazine_entries.index[0]
            magazine_community = G.nodes[magazine]['community']

            # Calculate topic score using scaled topic data
            topic_score = np.dot(df.loc[user_reviews.index, scaled_topic_columns].mean(), df.loc[magazine_index, scaled_topic_columns])
            
            # Community score
            community_score = 1 if user_community == magazine_community else 0

            # Combine the scores with community influence
            scores[magazine] = (
                weight_rating * average_rating + 
                weight_helpful_votes * total_helpful_votes + 
                weight_sentiment * average_sentiment_score + 
                weight_centrality * centrality_score + 
                weight_topic * topic_score + 
                weight_community * community_score  # Incorporate community matching into scoring
            )

    recommended_magazines = sorted(scores, key=scores.get, reverse=True)[:num_recommendations]
    return [magazine for magazine in recommended_magazines]


# # Example usage
# recommended_magazines = recommend_magazines('AE7Y5RLYIKHOZB5NKKOEKYG2SPSQ')
# print(recommended_magazines)
```

# Recommendation Experiment with different weight

```python


# Define a list of diverse configurations to test
weight_configurations = [
    {'rating': 0.4, 'helpful_votes': 0.2, 'sentiment': 0.2, 'centrality': 0.0, 'topic': 0.2, 'community': 0.0},   # Review and rating centric
    {'rating': 0.0, 'helpful_votes': 0.0, 'sentiment': 0.0, 'centrality': 0.6, 'topic': 0.0, 'community': 0.4},  # Only Network
    {'rating': 0.4, 'helpful_votes': 0.0, 'sentiment': 0.0, 'centrality': 0.3, 'topic': 0.0, 'community': 0.3},  #  Network and rating centric
    {'rating': 0.167, 'helpful_votes': 0.167, 'sentiment': 0.167, 'centrality': 0.167, 'topic': 0.167, 'community': 0.167}, # All balanced
    {'rating': 0.2, 'helpful_votes': 0.2, 'sentiment': 0.2, 'centrality': 0.1, 'topic': 0.2, 'community': 0.1}, # High emphasis on review itself
    {'rating': 0.1, 'helpful_votes': 0.1, 'sentiment': 0.1, 'centrality': 0.4, 'topic': 0.1, 'community': 0.2} # High emphasis on network
]
```

```python
# Function to run tests
def run_tests(weight_configurations, user_id='AE7Y5RLYIKHOZB5NKKOEKYG2SPSQ', num_recommendations=8):
    results = {}
    for index, weights in enumerate(weight_configurations):
        recommended_magazines = recommend_magazines(
            user_id=user_id,
            num_recommendations=num_recommendations,
            weight_rating=weights['rating'],
            weight_helpful_votes=weights['helpful_votes'],
            weight_sentiment=weights['sentiment'],
            weight_centrality=weights['centrality'],
            weight_topic=weights['topic']
        )
        results[f'Test {index + 1}'] = [meta_df[meta_df['parent_asin'] == i]['title'].iloc[0].replace("Print Magazine", "").strip() for i in recommended_magazines]
    return results

# Run the tests
test_results = run_tests(weight_configurations)
```

```python
def calculate_coverage(all_recommendations, total_items):
    unique_recommended_items = set(all_recommendations)
    return len(unique_recommended_items) / total_items

# Calculate total coverage
all_recommendations = sum(test_results.values(), [])  # Flatten the list of all recommended items
total_possible_items = meta_df['parent_asin'].nunique()  # Total unique magazines available in the dataset
total_coverage = calculate_coverage(all_recommendations, total_possible_items)
print("Total Coverage:",total_coverage)
# print(f"Total Coverage: {total_coverage:.2f}")

```

```python
# Heatmap of recommendation overlaps or similarities based on magazine attributes
def similarity_matrix(recommendations):
    matrix = np.zeros((len(recommendations), len(recommendations)))
    for i, config1 in enumerate(recommendations.keys()):
        for j, config2 in enumerate(recommendations.keys()):
            set1 = set(recommendations[config1])
            set2 = set(recommendations[config2])
            matrix[i, j] = len(set1.intersection(set2)) / len(set1.union(set2))  # Jaccard index for similarity
    return matrix

sim_matrix = similarity_matrix(test_results)
plt.figure(figsize=(10, 8))
sns.heatmap(sim_matrix, annot=True, xticklabels=test_results.keys(), yticklabels=test_results.keys())
plt.title("Jaccard Similarity Between Configurations")
plt.show()
```
